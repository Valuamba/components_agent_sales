```python
def get_assignee_and_reviewer(issue_id: str):
    url = f'https://youtrack.wildberries.ru/api/issues/{issue_id}?fields=$type,id,author,summary,customFields($type,id,projectCustomField($type,id,field($type,id,name)),value($type,avatarUrl,buildLink,color(id),fullName,id,isResolved,localizedName,login,minutes,name,presentation,text))'

    response = requests.get(url, headers=headers)
    data_dict = response.json()
    
    # Initialize variables to store the names
    creator_name = None
    assignee_name = None
    reviewer_name = None
    
    # Iterate through customFields to find the relevant fields
    for field in data_dict["customFields"]:
        field_name = field["projectCustomField"]["field"]["name"]
        if field_name == "Creator" and "value" in field:
            creator_name = field["value"].get("fullName")
        elif field_name == "Assignee" and "value" in field:
            assignee_name = field["value"].get("fullName")
        elif field_name == "Reviewer [HR]" and "value" in field:
            reviewer_name = field["value"].get("fullName")
    
    # Print the results
    print("Creator:", creator_name)
    print("Assignee:", assignee_name)
    print("Reviewer:", reviewer_name)

    return assignee_name, reviewer_name
```

```python
def get_priority(issue_id):
    url = f'https://youtrack.wildberries.ru/api/issues/{issue_id}?fields=$type,id,idReadable,author,summary,customFields($type,id,projectCustomField($type,id,field($type,id,name)),value($type,avatarUrl,buildLink,color(id),fullName,id,isResolved,localizedName,login,minutes,name,presentation,text))'

    response = requests.get(url, headers=headers)
    data_dict = response.json()
    
    priority_value = None
    for field in data_dict["customFields"]:
        if field["projectCustomField"]["field"]["name"] == "Priority":
            priority_value = field["value"]["name"]
            break
    
    # Print the Priority value
    print("Priority:", priority_value)
    return priority_value, data_dict['idReadable']
```

```python
def get_linked_issues(issue_id, connection_type=["Relates", "Depends", "Subtasks"]):
    get_issue_links_url = f'https://youtrack.wildberries.ru/api/issues/{issue_id}/links?fields=linkType(name,sourceToTarget,targetToSource),issues(id,idReadable,summary)'
    response = requests.get(get_issue_links_url, headers=headers)

    data_list = response.json()
    linked_issues = []
    for item in data_list:
        if item["linkType"]["name"] in connection_type:
            for issue in item.get("issues", []):
                issue_info = {
                    "id": issue.get("id"),
                    "idReadable": issue.get("idReadable")
                }
                linked_issues.append(issue_info)
    
    # Printing the result
    for issue in linked_issues:
        print(f'ID: {issue["id"]}, Readable ID: {issue["idReadable"]}')
    return linked_issues
```

```python
def get_issues_comments_authors(issue_id: str):
    get_issue_comments_url = f'https://youtrack.wildberries.ru/api/issues/{issue_id}/comments?fields=id,author%28login,name,id%29,deleted,text,updated,visibility%28permittedGroups%28name,id%29,permittedUsers%28id,name,login%29%29'

    response = requests.get(get_issue_comments_url, headers=headers)
    comments = response.json()
    
    authors = []
    for comment in comments:
       authors.append(comment['author']['name']) 
    
    authors = list(dict.fromkeys(authors))
    return authors
```

```python
youtrack_url = 'https://youtrack.wildberries.ru'
search_query = 'Business Line:HR   State: Resolved sort by: {Threat Score} desc'

response = requests.get(f'{youtrack_url}/api/issues?query={search_query}', headers=headers)

vulns_info = []
vuln_resolved_issues = response.json()
for vuln_issue in vuln_resolved_issues:
    vuln_id = vuln_issue['id']
    priority, readable_id = get_priority(vuln_id)
    vuln = {
        'id': vuln_issue['id'],
        'issue_id': readable_id,
        'priority': get_priority(vuln_id),
        'comments_authors': get_issues_comments_authors(vuln_id)
    }

    linked_issues = get_linked_issues(vuln_id)
    linked_issues_info = []
    for linked_issue in linked_issues:
        lissue_id = linked_issue['id']
        assignee, reviewer = get_assignee_and_reviewer(lissue_id)
        lissue_info = {
            'id': lissue_id,
            'idReadable': linked_issue['idReadable'],
            'comments_authors': get_issues_comments_authors(lissue_id),
            'assignee': assignee,
            'reviewer': reviewer
        }
        linked_issues_info.append(lissue_info)

    vuln['linked_issues'] = linked_issues_info
    vulns_info.append(vuln)

vulns_info
```

```python
def format_json(data):
    formatted_string = ""
    for issue in data:
        formatted_string += f"{issue['issue_id']} - {issue['priority'][0]}\n"
        formatted_string += f"Commented: {', '.join(issue['comments_authors'])}\n"
        if issue['linked_issues']:
            formatted_string += "Linked issues:\n"
            for linked_issue in issue['linked_issues']:
                formatted_string += f"    {linked_issue['idReadable']} - Assignee: {linked_issue['assignee']}, Reviewer: {linked_issue['reviewer']}\n"
                if 'comments_authors' in linked_issue and linked_issue['comments_authors']:
                    formatted_string += f"    Комментировали: {', '.join(linked_issue['comments_authors'])}\n"
        formatted_string += "\n"
    return formatted_string

# Pretty format the JSON data
pretty_formatted_json = format_json(vulns_info)
print(pretty_formatted_json)
```
