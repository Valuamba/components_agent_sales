```python
import pandas as pd

df = pd.read_json(r'C:\Users\MGroup\Documents\products.json')
```

```python

# Calculate the percentage change in price and quantity demanded for each item
df['price_sell'] = pd.to_numeric(df['price_sell'])
df['price_change'] = df.groupby('articul')['price_sell'].pct_change()
df['quantity_change'] = df.groupby('articul')['amount'].pct_change()

# Calculate Price Elasticity of Demand
df['PED'] = df['quantity_change'] / df['price_change']

# Display the results
df.dropna(inplace=True)  # Remove rows with NaN values (first rows for each articul where pct_change cannot be calculated)
df
```
