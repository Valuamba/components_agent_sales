```python
import re
from bs4 import BeautifulSoup


with open('./deals_html/discount_v4/447713_41039.0.html', 'r') as f:
    html = f.read()


# Parse the HTML content
soup = BeautifulSoup(html, "html.parser")

# Get all text content from the HTML
text_content = soup.get_text(separator="\n")

# Clean the text content by replacing multiple newlines and spaces with single spaces
cleaned_text = re.sub(r'\s+', ' ', text_content)

# Find all positions of "From:" lines
from_positions = [match.start() for match in re.finditer(r'From:', cleaned_text)]

# Define the required segments
required_segments = ["From:", "Date:", "To:", "Subject:"]

# Define a function to check if a block contains the required email metadata
def is_email_metadata_block(block_text):
    return all(segment in block_text for segment in required_segments)

# Initialize the counter and list for valid blocks
email_metadata_blocks = []

# Iterate over positions to extract blocks
for i, start_pos in enumerate(from_positions):
    # Determine the end position
    end_pos = from_positions[i + 1] if i + 1 < len(from_positions) else len(cleaned_text)
    
    # Extract block text
    block_text = cleaned_text[start_pos:end_pos]
    
    # Check if the block contains the required email metadata
    if is_email_metadata_block(block_text):
        email_metadata_blocks.append(block_text)

# Output the count of matches and the matched blocks for verification
email_metadata_block_count = len(email_metadata_blocks)
print(f"Found {email_metadata_block_count} email metadata blocks.")
```

```python
df['count_from'] = df['plain_text_body'].apply(lambda row: row.count('From:') if isinstance(row, str) else 0)
df['count_offer_nr'] = df['plain_text_body'].apply(lambda row: row.count('Offer-Nr.:') if isinstance(row, str) else 0)
df['count_forwarded'] = df['plain_text_body'].apply(lambda row: row.count('-------- Forwarded Message --------') if isinstance(row, str) else 0)
```
