# Messages filtered by intents and from

````python
import json


# Load your deals.json file
with open('./assets/deals.json', 'r') as file:
    deals_data = json.load(file)

specified_intents = similar_intents[:2]

# Function to check if a message's intent is in the specified intents
def is_intent_in_specified_intents(message_intents):
    for intent in message_intents:
        if intent in specified_intents:
            return True
    return False

def is_message_already_added(selected_messages, message):
    return any(msg['id'] == message['id'] for msg in selected_messages)

# List to store the filtered deals and their messages
filtered_deals = []

# Iterate over each deal and its messages
for deal in deals_data['deals']:
    selected_messages = []
    for i, message in enumerate(deal['messages']):
        if message['from'] == 'customer' and is_intent_in_specified_intents(message['intents']):
            # Add the previous message if it exists and is not already added
            if i > 0 and not is_message_already_added(selected_messages, deal['messages'][i - 1]):
                selected_messages.append(deal['messages'][i - 1])

            # Add the current message if it's not already added
            if not is_message_already_added(selected_messages, message):
                selected_messages.append(message)

            # Add the next message if it exists and is not already added
            if i < len(deal['messages']) - 1 and not is_message_already_added(selected_messages, deal['messages'][i + 1]):
                selected_messages.append(deal['messages'][i + 1])

    if selected_messages:
        filtered_deals.append({"deal_id": deal["deal"]["deal_id"], "selected_messages": selected_messages})

# Print or process the filtered deals and their messages
# print(filtered_deals)
print(json.dumps(filtered_deals, indent=2))
````

````python
def format_deals_to_string(deals):
    formatted_string = ""
    
    for deal_index, deal in enumerate(deals, start=1):
        formatted_string += f"Deal #{deal_index}\n\n"
        
        for msg_index, message in enumerate(deal['selected_messages']):
            sender = message['from']
            body = message['body']
            intents = "\n".join([f"    {intent['intent']} -> {intent['sub_intent']} -> {intent['branch']}" 
                                 for intent in message['intents']])
            
            formatted_string += f"Msg: {msg_index}\n"
            formatted_string += f"{sender}: {body}\n"
            formatted_string += "Intents:\n" + intents + "\n\n"

    return formatted_string

# Assuming 'deals_data' is your JSON data
deals_chat_histories_filtered_by_intents = format_deals_to_string(filtered_deals)
print(deals_chat_histories_filtered_by_intents)
````
